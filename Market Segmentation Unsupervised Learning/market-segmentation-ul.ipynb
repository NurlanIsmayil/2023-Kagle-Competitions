# Intoduction about MARKET SEGMENTATION?


Market segmentation is a vital concept in the realm of marketing, denoting the strategic procedure of subdividing an expansive market, encompassing both existing patrons and prospective clients, into distinct subsets of consumers. This segregation is grounded in the identification of common attributes, behaviors, or characteristics that unite individuals within each subgroup, facilitating the development of more targeted and effective marketing strategies.

In essence, market segmentation is akin to dissecting a heterogeneous market into more manageable and homogenous pieces, enabling businesses to better understand the unique needs, preferences, and demands of specific consumer clusters. This facilitates the creation of tailor-made marketing campaigns, products, and services that resonate with the particular interests and behaviors of each subgroup.

Market segmentation serves as a powerful tool for businesses seeking to enhance their market penetration, boost customer engagement, and optimize resource allocation. By recognizing the diverse dynamics at play within a broader market and categorizing consumers into meaningful segments, companies can craft personalized marketing messages and experiences, leading to higher customer satisfaction and brand loyalty.

In conclusion, market segmentation is the strategic act of dividing a diverse consumer or business market into smaller, more homogeneous groups based on shared characteristics, a fundamental strategy that empowers businesses to meet the distinct needs of their target audiences, fostering stronger customer relationships and overall market success.

# Now Import Some More Important Libs

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN,SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import metrics
```

#  Now Scaling the DataFrame and Dimensionality reduction

```python
scl=StandardScaler()
scaled_df = scl.fit_transform(df)
```

```python
# Dimensionality reduction-Converting the DataFrame into 2D DataFrame for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

# Hyperparameter tuning 


```python
# Finding 'k' value by Elbow Method
inertia = []
range_val = range(1,10)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

# Model Building using KMeans

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

# Visualizing the clustered dataframe

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

# To find all cluster centers

```python
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
```

# Training and Testing the model accuracy using decision tree

```python
#Split Dataset
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3,random_state=3)
```

```python
X_train
```

```python
X_test
```

# Decision Tree Classifier

```python

model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#Confusion_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
